## **Introducción**

El análisis cinemático y dinámico de manipuladores robóticos constituye una herramienta fundamental para comprender y controlar su comportamiento, así como ser uno de los conocimientos bases en la formación del profesional en el campo de la mecatrónica. Dicho conocimiento se debe ver ver reflejado en la aplicación de los conocimientos de la física y matemáticas adquiridos a lo largo del aprendizaje universitario y desde antes. 

En el trabajo expuesto se desarrolla el modelo de un robot SCARA de tres eslabones abordando cinemáticas y la dinámica, en ambos casos de forma directa e inversa. En la primera parte se aborda el modelo cinemático de la postura, incluyendo el cálculo de vector de posiciones y se discute la solución del modelo imverso. 

La segunda y tercera parte, se estudia y se desarrolla el modelado cinemático de las velocidades y aceleraciones respectivamente, mostrando la relación entre las variables articulares y las del punto de interés del efector final. El desarrollo de estas secciones muestran como las ecuaciones de velocidades y aceleraciones proyectan las contribuciones tangenciales y normales de cada articulación sobre el sistema de referencia inercial; obteniendo de forma sistemática la velocidad y aceleración de cada eslabón, acumulando los efectos desde la base del robot hasta el punto P de interés. Para el modelo cinemático inverso de las aceleraciones, se definió un vector de aceleraciones representativas con el fin de mostrar el procedimiento del cálculo del modelo inverso.

En la cuarta parte se aborda el modelo dinámico mediante las ecuaciones de Euler-Lagrange; se detallan los pasos para la obtención de la matriz de inercia, vectores de la fuerza de Coriolis, centrífugas y de la gravedad; para construir el modelo dinámico directo. A partir de este desarrollo se obtiene el modelo dinámico inverso que permite calcular aceleraciones articulares o fuerzas de restricción en función de los torques aplciados. En esta parte, las ecuaciones se estructuran de manera análoga a las proyecciones de velocidades y aceleraciones en cinemática, cada térmico representa una proyección distinta. Así pues, el modelo dinámico integra todas las fuerzas y energías desde la base del robot hasta el putno P de interés.
