## Previo de la práctica

- #### **Estructura de paquetes en ROS 2** [1], [2]

En ROS2, un paquete es la unidad básica de organización, el cual contiene:

- Código fuente
- Archivos de configuración
- Dependencias
- Metadatos

La estructura típica de un paquete en ROS2 suele tener la siguiente organización:

- package.xml:
    - Archivo de metadatos: nombre del paquete, versión, licencias, dependencias.
    - Obligatorio en todos los paquetes.

- CMakeList.txt ó setup.py:
    - Define como se compila o instala el paquete
    - `CMakeList.txt` para paqeuetes en C++
    - `setup.py` para paquetes en Python

- Carpetas principales
    - `src/` : código fuente (nodos, librerías).
    - `include/` : cabeceras C++.
    - `resource/` : archivos de soporte, datos, configuraciones.
    - `launch/` : archivos de lanzamiento (launch.py) para iniciar nodos.
    - `test/` : pruebas unitarias o de integración.

- Archivos adicionales 
    - `README.md` : docuemntación del paquete.
    - `setup.cfg` : configuración de instalación de python.
    - `COLCON_IGNORE` : indica a colcon que ignore el paquete.

El flujo de creación de un paquete se muestra a continuación:

1. Creación del workspace
    - `mkdir -p ~/ros2_ws/src`
    - `cd ~/ros2_ws/src`

2. Generación del paquete
    - Para C++: 
        - `ros2 pkg create --build-type ament_cmake my_cpp_pkg`
    - Para python:
        - `ros2 pkg create --build-type ament_python my_py_pkg`

3. Compilar con colcon
    - `cd ~/ros2_ws`
    - `colcon build`
    - `source install/setup.bash`

- #### **Definición de nodos y tópicos** [3], [4]

En ROS 2, los ***nodos*** son procesos que ejecutan código y representan las unidades de computación, mientras que los ***tópicos*** son canales de comunicación que le permiten a los nodos intercambiar mensajes de manera asíncrona. Los nodos pueden publicar en múltiples tópicos y suscribirse a otros tantos. 

**Nodo**

Es una instancia de un proceso que realiza una tarea específica dentro del sistema. Dentro de sus características prencipales se tienen:
- Cada nodo tiene nombre único dentro del  grafo ROS
- Puede ejecutar funciones como publicar, suscribirse, ofrecer servicios o usar parámetros.
- Se gestionan con comandos como `ros2 run`, `ros2 node list`, `ros2 node info`.


**Tópico**

Es un bus de comunicación que conecta nodos mediante mensajes. Dentro de sus características principales se encuentran:
- Son asíncronos: los publicadores no necesitan saber quien está suscrito.
- Cada tópico tienen un nombre y tipo de mensaje asociado.
- Tiene herramientas útiles tales como:
    - `ros2 topic list` -> muestra los tópicos activos
    - `ros2 topic echo <topic>` -> imprime los mensajes recibidos
    - `ros2 topic pub <topic>` -> publica manualmente un tópico

**Resumen**

|**Concepto**|*Nodo*|*Tópico*|
|:------:|:--:|:----:|
|*Qué es*|Proceso que ejecuta código|Canal de comunicación|
|*Función*|Computación (publicar, suscribir, servicios)|Transporte de mensajes|
|*Relación*|Los nodos usan tópicos para comunicarse|Los tópicos existen si hay nodos publicando o suscribiendo|
